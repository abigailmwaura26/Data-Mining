```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime, timedelta

# Define tasks and their durations
tasks = {
    "Project Initiation": (datetime(2024, 1, 1), datetime(2024, 2, 15)),
    "AI Algorithm Development": (datetime(2024, 2, 1), datetime(2024, 7, 31)),
    "Security & Compliance Integration": (datetime(2024, 3, 1), datetime(2024, 6, 30)),
    "User Interface Design": (datetime(2024, 3, 15), datetime(2024, 7, 15)),
    "System Testing & QA": (datetime(2024, 6, 1), datetime(2024, 8, 15)),
    "Pilot Deployment": (datetime(2024, 8, 1), datetime(2024, 10, 31)),
    "Marketing & Branding": (datetime(2024, 5, 1), datetime(2024, 10, 15)),
    "Sales Team Onboarding": (datetime(2024, 6, 1), datetime(2024, 9, 30)),
    "Operational Readiness": (datetime(2024, 7, 1), datetime(2024, 10, 31)),
    "Full Market Launch": (datetime(2024, 11, 1), datetime(2024, 12, 31))
}

# Prepare data for Gantt chart
task_names = list(tasks.keys())
start_dates = [task[0] for task in tasks.values()]
end_dates = [task[1] for task in tasks.values()]
durations = [(end - start).days for start, end in zip(start_dates, end_dates)]

# Plot the Gantt chart
fig, ax = plt.subplots(figsize=(12, 8))
ax.barh(task_names, durations, left=start_dates, color='skyblue')
ax.set_xlabel('Timeline')
ax.set_title('Gantt Chart for Cognicare Implementation')

# Format the x-axis to display dates
ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))
plt.xticks(rotation=45)

# Add grid lines and adjust layout
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()

# Display the Gantt chart
plt.show()
```
